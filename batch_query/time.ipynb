```python
from balif import BADIForest, odds_datasets
from balif.active_learning import get_queries_independent, get_queries_worstcase, get_queries_batchbald

for dataset in odds_datasets.datasets_names:
    X, y = odds_datasets.load(dataset)
    model = BADIForest(contamination=y.mean()).fit(X)
    print(f"\nDATASET:     {dataset[:10]}\tX: {X.shape}\tcontamination: {100*y.mean():.2f}%")
    for batch_size in [10]:
        print(f"\nBATCH SIZE: {batch_size}")

        # RANDOM
        print("random          \t", end="")
        %timeit -n 10 model.decision_function(X)

        # INDEPENDENT 
        for interest_method in ["bald", "margin", "anom"]:
            print(f"independent {interest_method}\t", end="")
            %timeit -n 10 get_queries_independent(model, X, interest_method, batch_size)

        # WORSTCASE
        for interest_method in ["bald", "margin", "anom"]:
            print(f"worstcase {interest_method}  \t", end="")
            %timeit -n 10 get_queries_worstcase(model, X, interest_method, batch_size)

        # BATCHBALD
        print(f"batchbald       \t", end="")
        %timeit -n 10 get_queries_batchbald(model, X, batch_size)
```
